### Objetivo: Identificar elementos de una visualización basados en el framework de Tamara

La siguiente imágen proviene del artículo ***The Words Men and Women Use When They Write About Love*** de ***The New York Times***

***¿Qué?***

* Se pueden observar palabras que se usan con mayor y menor frecuencia en ensayos publicados e inéditos según el género del escritor.
* Estan presentes dos variables categóricas, las cuales son el género (Masculino o Femenino) y el tipo de ensayo (Publicado o Inédito).
* La información que se presenta es de tipo ordenada ordinal dado que se mide la frecuencia del uso de palabras específicas y existen palabras con distintas frecuencias.
* Los datos que se presentan son de tipo tabla.
* Se presentan datos estáticos. 

***¿Por qué?***

* Al momento de usar este gráfico de burbujas, permite comparar tendencias del uso de palabras específicas en ensayos del New York Times entre mujeres y hombres. 
* Identificar frecuencias de las palabras con mayor y menor uso en ensayos según el género.

***Marcas y canales utilizados***

* Marca de puntos
* Canal de posición espacial horizontal y vertical para los atributos categóricos (género y tipo de ensayo).
* Canal de color para el atributo categórico del genero. 
* Canal de tamaño para el atributo cuantitativo (palabras con mayor o menor uso)

***Cumple o no cumple alguna de las reglas generales descritas***

* Una de las reglas que podría cumplir esta visualización es "Los ojos le ganan a la memoria", dado que puede ser fácil para el usuario retener en la memoria las palabras que son más frecuentes en hombres y mujeres, sin embargo pueden presentar dificultades al recordar la mayoria de la información que se presenta. 
* Una de las reglas que no cumple esta visualización es "Resolución sobre la inmersión", en este caso es posible que al momento de leer las palabras con menor uso en los ensayos se le dificulte al usuario posiblemente por la cantidad de pixeles que tiene el dispositivo donde observa el gráfico. 
* Esta visualización no cumple con las reglas: "Primero overview, zoom y filtro, detalles a petición" y "Primero la función, luego la forma"; para el usuario resulta complicado acceder a las palabras con menor uso, esto sucede por la sobreposición de las burbujas lo que hace que el usuario se deba esforzar para poder identificar la información y se pierde el objetivo de la visualización. 

***Mejoras posibles en cuanto a marcas o canales, ¿qué le realizaría a la visualización?***

Clasificaría la información según el tipo de ensayo (variable categórica), posteriormente haría dos nubes de burbujas manteniendo el tamaño y los colores que ayudan a identificar el género (variable categórica) pero aumentando y disminuyendo la intensidad del color, teniendo en cuenta la frecuencia de uso de la palabra, así las palabras con una burbuja grande tendrán una intensidad mayor que una burbuja pequeña, y finalmente evitaria que se sobrepongan las burbujas para que el usuario pueda acceder a toda la información más facilmente. 

La siguiente imágen tiene como título ***Because every country is the most beautiful at something*** esta visualización fue publicada por ***information is beautiful***

***¿Qué?***

* Se puede observar que en la parte superior del gráfico existe una clasificación por temas y colores, algunos paises tienen una leyenda que indica en qué se resalta según la categoría en la que esté. 
* La información que se presenta es estática; en este caso los paises tienen un color particular que les fue asignado según la variable categórica "razones para celebrar". 
* Los datos que se presentan son de tipo tabla. 

***¿Por qué?***

* Presentar e identificar características de todos los paises del mundo. 
* Localizar atributos similares entre los paises del mundo. 



***Marcas y canales utilizados***

* Marca de areas usada en cada país.
* Canal de posición espacial horizontal y vertical tanto en los países como en ls atributos de cada uno.
* Canal de color para la variable categórica "razones para celebrar". 

* Existe eficiencia del canal, dado que se puede observar la "separabilidad" y la "agrupación" entre los paises por medio de los colores. 

***Cumple o no cumple alguna de las reglas generales descritas***

* El gráfico no cumple con la regla "Primero la función, luego la forma", ya que para el usuario puede ser complicado entender qué característica corresponde a qué país, además la leyenda de las categorías no resalta a simple vista, por lo que es necesario buscar dentro del gráfico si los colores representan alguna relación entre los paises. 
* El gráfico cumple con la regla "2D injustificado", la información que se presentan en el gráfico se podría organizar en listas o tablas, así la información estaría más clara y no existiría pérdida de información, pues al tener unos paises más pequeños que otros la ubicación de las características quedan por fuera del mapa creando confusión o en otros casos no se percibe a simple vista, es necesario seleccionar el país pequeño con el cursor para observar la característica de este. 

***Mejoras posibles en cuanto a marcas o canales, ¿qué le realizaría a la visualización?***

* Se podría mantener el mapa, sin embargo, le cambiaría la paleta de colores por unos que contrasten entre ellos y así poder resaltar mejor cada categoría; también cambiaría la forma de presentar la información, es decir, no dejaría toda la información explícita y estática, lo mejoraría permitiendo al usuario seleccionar la categoría de interés y haría que el mapa solo habilitara la categoría seleccionada, sólo los paises que estén dentro de esa categoría tomarían color y cada uno mostraría en lo que se destaca, evitando la saturación de información. 

Para el caso de los paises pequeños, le pondría unas viñetas o unas flechas a la información en la que se destaca, evitando que la leyenda quede "suelta" o sin asignar, evitaría poner todas las leyendas al mismo tiempo, inicialmente se podría dejar todo el mapa con colores pero sin las leyendas, pues estas se activarian una vez el usuario seleccione la categoría que le interese y resaltaría la leyenda de "razones para celebrar" para que el usuario entienda con mayor facilidad el gráfico. 

* Si cambio el mapa por otro tipo de visualización, agruparía los paises que estan dentro de una categoría específica y los dejaría con un color uniforme, resaltando en cada uno la razón por la que se destaca en la categoría.
